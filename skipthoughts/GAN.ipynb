```python
def relu(x):
    return tf.nn.relu(x)

class deconv_model():
    def __init__(self):
        n1, n2, n3 = 16, 32, 64
        ksize = 5
        self.input_vect = tf.placeholder(tf.float32, shape=(None, 4800))
        #self.input_expanded = tf.expand_dims(self.input_vect, 1)
        self.W_in = tf.Variable(tf.random_normal([4800, 400], stddev=.1))
        self.in_fc_linear = relu(tf.matmul(self.input_vect, self.W_in))
        self.in_fc = tf.reshape(self.in_fc_linear, (1, 20, 20, 1))
        self.W_conv_1 = tf.Variable(tf.random_normal([ksize, ksize, 1, n1], stddev=.1))
        self.dec_1 = tf.nn.conv2d(self.in_fc, self.W_conv_1, strides=[1,2,2,1], padding='SAME')
        self.session = tf.Session()
        self.session.run(tf.initialize_all_variables())
    def run(self, a):
        out = self.session.run(self.dec_1, {self.input_vect: a})
        print(out)
```

```python

dec = deconv_model()
dec.run([a])
```
